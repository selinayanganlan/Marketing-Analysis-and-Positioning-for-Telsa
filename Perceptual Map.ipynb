# Data Cleaning

```python
import pandas as pd
data = pd.read_excel('Customer Perception on Electric Vehicles_May 22, 2021_17.54.xlsx')  # 或xls
pd.set_option('display.max_columns', None)  # 不隐藏列
data.head()
```

```python
data.shape
```

```python
data.drop(data.index[0], inplace=True)
data.head()
```

```python
data['Q1_1'].value_counts()
```

```python
data.columns.get_loc('Q1_1')
```

```python
data.columns.get_loc('Q5_7')
```

```python
survey_data = data.iloc[:, 17:45]
survey_data.head()
```

```python
for i in survey_data.columns:
    survey_data[i] = survey_data[i].apply(lambda x: 1 if x == '1 - Extremely Disagree'
                                            else 5 if x == "5 - I'm not sure"
                                            else 10 if x == '10 - Extremely Agree'
                                            else x)
survey_data.head()
```

```python
survey_data.info()
```

```python
for i in survey_data.columns:
    survey_data[i] = survey_data[i].astype('int')

survey_data.info()
```

```python
survey_data_normalized = survey_data.copy()

for i in survey_data_normalized.columns:
    survey_data_normalized[i] = survey_data_normalized[i].apply(lambda x: x-5)
    
survey_data_normalized.head()
```

```python
# 加权平均
survey_data_normalized['Q1_F1'] = 0.457*survey_data_normalized['Q1_3'] + 0.957*survey_data_normalized['Q1_5'] + 0.528*survey_data_normalized['Q1_6'] + 0.503*survey_data_normalized['Q1_7']
survey_data_normalized['Q1_F2'] = 0.555*survey_data_normalized['Q1_2'] + 1.042*survey_data_normalized['Q1_4'] + 0.402*survey_data_normalized['Q1_6']

survey_data_normalized['Q2_F1'] = 0.457*survey_data_normalized['Q3_3'] + 0.957*survey_data_normalized['Q3_5'] + 0.528*survey_data_normalized['Q3_6'] + 0.503*survey_data_normalized['Q3_7']
survey_data_normalized['Q2_F2'] = 0.555*survey_data_normalized['Q3_2'] + 1.042*survey_data_normalized['Q3_4'] + 0.402*survey_data_normalized['Q3_6']

survey_data_normalized['Q3_F1'] = 0.457*survey_data_normalized['Q4_3'] + 0.957*survey_data_normalized['Q4_5'] + 0.528*survey_data_normalized['Q4_6'] + 0.503*survey_data_normalized['Q4_7']
survey_data_normalized['Q3_F2'] = 0.555*survey_data_normalized['Q4_2'] + 1.042*survey_data_normalized['Q4_4'] + 0.402*survey_data_normalized['Q4_6']

survey_data_normalized['Q4_F1'] = 0.457*survey_data_normalized['Q5_3'] + 0.957*survey_data_normalized['Q5_5'] + 0.528*survey_data_normalized['Q5_6'] + 0.503*survey_data_normalized['Q5_7']
survey_data_normalized['Q4_F2'] = 0.555*survey_data_normalized['Q5_2'] + 1.042*survey_data_normalized['Q5_4'] + 0.402*survey_data_normalized['Q5_6']

survey_data_normalized.head()
```

```python
Q1_F1 = survey_data_normalized['Q1_F1'].mean()
Q1_F2 = survey_data_normalized['Q1_F2'].mean()

Q2_F1 = survey_data_normalized['Q2_F1'].mean()
Q2_F2 = survey_data_normalized['Q2_F2'].mean()

Q3_F1 = survey_data_normalized['Q3_F1'].mean()
Q3_F2 = survey_data_normalized['Q3_F2'].mean()

Q4_F1 = survey_data_normalized['Q4_F1'].mean()
Q4_F2 = survey_data_normalized['Q4_F2'].mean()
```

```python
df = pd.DataFrame({"Car": ['Tesla', 'BMW', 'Hyundai', 'Audi'],
                                         "F1": [Q1_F1, Q2_F1, Q3_F1, Q4_F1],
                                         "F2": [Q1_F2, Q2_F2, Q3_F2, Q4_F2]})
df
```

# Perceptual Map

```python
import plotly.express as px

fig = px.scatter(df, x="F1", y="F2", hover_name="Car", text='Car',
                labels=dict(F1="Moderate Performance vs High Performance", F2="Traditional vs Technology Driven"))

fig.update_xaxes(range=[-6,6])
fig.update_yaxes(range=[-6,6])
fig.update_traces(textposition='top center')

fig.update_layout(
    title = {
        'text': "Perceptual Map",
      # 'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}
)

fig.show()
```
